### trabalhando com requests (método GET)

```python
#imports
import requests
from bs4 import BeautifulSoup
```

```python
http_proxy  = "http://10.130.214.242:3128"
https_proxy = "https://10.130.214.242:3128"
ftp_proxy   = "ftp://10.130.214.242:3128"

proxyDict = { 
              "http"  : http_proxy,
              "https" : https_proxy,
              "ftp"   : ftp_proxy
            }
```

```python
url = 'https://www.youtube.com/results?'
payload = {'search_query':'yamanha r'}
```

```python
r = requests.get(url, proxies=proxyDict, params=payload)
```

```python
#r.content
r.url
```

```python
#salvando conteudo da pagina no arquivo
with open('youtube.html', 'w') as file:
    file.write(r.text)
```

### metodo POST (correios)

```python
url_correios = 'http://www.buscacep.correios.com.br/sistemas/buscacep/resultadoBuscaCepEndereco.cfm'
```

```python
payload_pesquisa = {
    'relaxation':'19020600',
    'tipoCEP': 'ALL',
    'semelhante': 'N'
}
```

### parâmetros requests.post(url, data=''):
    ** data= (dados) para pesquisa (via POST)
    ** params= (parametro de busca via URL/ GET)

```python
response_correios = requests.post(url_correios, proxies=proxyDict, data=payload_pesquisa) #data (dados) da pesquisa
```

```python
#response_correios.text
c = BeautifulSoup(response_correios.content, 'lxml')
cep_result = c.find('table', class_='tmptabela')
```

```python
listc = []
for string in cep_result.stripped_strings:
    listc.append(string)
    print(repr(string))
```

```python
dictc={}
```

```python
for i in range(0,4,1):
    dictc[listc[i]] = listc[i+4]
```

### Erros e Exceções

```python
url = 'http://www.google.com'
```

```python
try:
    r = requests.get(url, proxies=proxyDict, timeout=2)
    print('try')
except requests.exceptions.ConnectionError as e: #geralmente porque o site non ecziste
    print('Connection\n' + str(e))
except requests.exceptions.Timeout as t:
    print('Erro de Timóteo\n' + str(t))
except requests.exceptions.TooManyRedirects as m: #erro de muitas requisicoes (acima de 5)
    print('TooManyRedirects\n' + str(m))
except requests.exceptions.RequestException as re: #URL inexistente
    print('RequestException\n' + str(re))
```
